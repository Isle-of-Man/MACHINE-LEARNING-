#### Importance of w0 

details about wo
- its the distance of the line from the center. 
- it is directly proportional to the negative of the distance

- w1x1 + w2y1 + wo = 0 ie w2y1 = - w1x1 - wo 
- y1 = -w1/w2 x1 - wo/w2 
- therefore proving the intercept is - wo/w2 and the distance is proportional to -wo

Purpose of Wo and W vector (wT)
- w1x1 + w2y1 + wo = 0 for this quuation -
- wT ie W vector is [ w1,w2,....wn] - plot these and it wont be in the line but this wT will give us the direction perpendicular to the decision plane 
- wo will move the decision plane in the direction of wT. 

Distance of a h-plane from the origin
- -wo/||w||
- ||w|| = sqrt(x^2 + y^2)

Distance of a point (xo) from a h-plane
- (wT*xo + wo)/||w||
